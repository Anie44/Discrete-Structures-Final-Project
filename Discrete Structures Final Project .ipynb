```python
def create_grid(rows, columns):
    grid = [[' ' for _ in range(columns)] for _ in range(rows)]
    return grid

def print_grid(grid, braces):
    print("Welcome to Kam's Grid Bracing Analyser")
    print("How many rows?".ljust(25), len(grid))
    print("How many columns?".ljust(25), len(grid[0]))
    
    # Print column labels
    print(" " * 6 + " ".join(chr(65 + i) for i in range(len(grid[0]))))
    
    # Print top border
    print(" " * 6 + "+-----" * len(grid[0]) + "+")
    
    # Print grid rows
    for i, row in enumerate(grid):
        print(" " * 4 + chr(65 + i) + " |", end="")
        for j, cell in enumerate(row):
            if (i, j) in braces:
                print("\\", end=" " * 5)  # Adjust spacing here
            else:
                print(" " * 5, end="")
            print("|", end="")
        print("\n" + " " * 6 + "+-----" * len(grid[0]) + "+")

def add_brace(grid, braces):
    for brace in braces:
        row, col = brace
        if row < len(grid) and col < len(grid[0]):
            grid[row][col] = "\\"
        else:
            print(f"Brace {chr(row + 65)}{col + 1} is outside the grid. Ignoring.")

def calculate_rigidity(grid, braces):
    rigidity = 0
    for brace in braces:
        row, col = brace
        if row < len(grid) - 1 and col < len(grid[0]) - 1:
            rigidity += 1
        if row > 0 and col > 0:
            rigidity += 1
    return rigidity

def main():
    rows = int(input("How many rows? "))
    columns = int(input("How many columns? "))

    grid = create_grid(rows, columns)
    print("How many bracings do you want to add?")
    num_bracings = int(input())
    braces = []
    for i in range(num_bracings):
        brace_input = input(f"Enter brace {i+1} position (e.g., AF): ").upper()
        braces.append((ord(brace_input[0]) - 65, ord(brace_input[1]) - 65))

    add_brace(grid, braces)

    print_grid(grid, braces)

    rigidity = calculate_rigidity(grid, braces)
    print("\nThe structural rigidity of the grid bracing:", rigidity)

if __name__ == "__main__":
    main()
```
